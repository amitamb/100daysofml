```python
import numpy as np
import math
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def dx_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

class NNNot:
    def __init__(self):
        self.w1 = 5
        self.b1 = -2
        self.z1 = 0
        self.o1 = 0
    
    def forward(self, ival, oval):
        self.z1 = ival * self.w1 + self.b1
        print("z1 = %d * %f + %f = %d" % (ival, self.w1, self.b1, self.z1))
        self.a1 = sigmoid(self.z1)
        print("a1 = sigmoid(z1) = sigmoid(%f) = %f" % (self.z1, self.a1))
        a1_error = (self.a1 - oval) ** 2
        print("a1_error = (self.a1 - oval) ** 2 = (%f - %f) ** 2 = %f" % (self.a1, oval, a1_error))
        
        dx_z1 = dx_sigmoid(self.z1)
        print("dx_z1 = %f" % dx_z1)
        
        z1_error = dx_z1 * a1_error
        print("z1_error = dx_z1 * a1_error = %f * %f = %f" % (dx_z1, a1_error, z1_error))
        
        w1_error = ival * z1_error
        print("w1_error = ival * z1_error = %f * %f = %f" % (self.w1, z1_error, w1_error))
        
        b1_error = ival * z1_error
        print("b1_error = ival * z1_error = %f * %f = %f" % (self.b1, z1_error, b1_error))
        
        print("new w1 = old w1 - w1_error = %f - %f = %f" % (self.w1, w1_error, (self.w1 - w1_error)))
        self.w1 = self.w1 - w1_error
        
        print("new b1 = old b1 - b1_error = %f - %f = %f" % (self.b1, b1_error, (self.b1 - b1_error)))
        self.b1 = self.b1 - b1_error

#     def backward(self):
#         
```
